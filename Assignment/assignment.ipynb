```python
merge_data1=pd.concat([orderdetails,cookingsessions], axis=1)
```

```python
merge_data2=merge_data1.T.drop_duplicates()
```

```python
merge_data2.head(5)
```

```python
merge_data3=merge_data2.T
```

```python
merge_data3.head()
```

```python

merge_data = pd.merge(merge_data3,UserDetails, on="User ID", how="left")
```

```python
merge_data.head(20)
```

```python
merge_data.T.duplicated()
```

```python
no_order_dish= merge_data['Dish Name'].value_counts()
```

```python
no_order_dish.head()
```

```python
import matplotlib.pyplot as plt
```

```python
no_order_dish.plot(kind='bar', title="Top Popular Dishes based on order")
plt.xlabel("Dist Name")
plt.ylabel("Number of Orders")
plt.show()
```

```python
merge_data['Order Status'].value_counts()
```

```python
(merge_data['Meal Type'].value_counts()).plot(kind='bar',title='No. of Order in Different Type of Meal  ')
plt.xlabel('Type of meal')
plt.ylabel('Number of Orders')
```

```python
merge_data['Time of Day'].value_counts().plot(kind='pie', autopct='%1.1f%%', title='No of Orders in Different Time of Day')
plt.ylabel('')
plt.show()
```

```python
behavior_by_age = merge_data.groupby('Age')['Total Orders'].mean()
print(behavior_by_age)
```

```python
(merge_data['Age'].value_counts()).plot(kind='bar',title='No. of Order by Age')
plt.xlabel("Age group")
plt.ylabel('number of Order')
```

```python
cooking_order_relation = merge_data.groupby('User ID')[['Session ID', 'Order ID']].sum()
correlation = cooking_order_relation.corr()
print(correlation)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,7))
plt.scatter(merge_data['Session ID'], merge_data['Order ID'])
plt.title('Cooking Sessions vs Orders')
plt.xlabel('Cooking Sessions')
plt.ylabel('Orders')
plt.show()
```
